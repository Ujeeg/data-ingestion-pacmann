# Create Table CampaignItems

```python
df_excel["campaign_item_id"] = df_excel["campaign_objects"].str.split(";")
```

```python
df_excel_explode = df_excel.explode("campaign_item_id")
```

```python
df_excel_explode = df_excel_explode[["campaign_item_id","campaign_id","campaign_type"]]
```

```python
df_excel_explode["campaign_type"] = df_excel_explode["campaign_type"].astype(str)
```

```python
df_excel_explode.head()
```

```python
cond = df_excel_explode["campaign_type"] == "Category"
df_category_campaigns = df_excel_explode[cond]
```

```python
cond = df_excel_explode["campaign_type"] == "Product"
df_product_campaigns = df_excel_explode[cond]
```

```python
df_category_campaigns.loc[:, "campaign_item_id"] = df_category_campaigns["campaign_item_id"].astype(int)
df_categoty_finale["category_id"] = df_categoty_finale["category_id"].astype(int)
df_merge_category = df_category_campaigns.merge(
    df_categoty_finale, left_on="campaign_item_id", right_on="category_id", how="inner"
)
```

```python
df_product_campaigns.loc[:, "campaign_item_id"] = df_product_campaigns["campaign_item_id"].astype(int)
df_product_finale["product_id"] = df_product_finale["product_id"].astype(int)
df_merge_product = df_product_campaigns.merge(
    df_product_finale, left_on="campaign_item_id", right_on="product_id", how="inner"
)
```

```python
df_merge_product.head()
```

```python
df_merge_category = df_merge_category[["campaign_item_id","campaign_id","campaign_type","name"]]
df_merge_product = df_merge_product[["campaign_item_id","campaign_id","campaign_type","name"]]
```

```python
df_CampaignItems = pd.concat([df_merge_category, df_merge_product], ignore_index=True)
df_CampaignItems = df_CampaignItems.rename(columns={'name': 'campaign_item'})  
```

```python
import pandas as pd
import os

def save_df_to_csv(df, directory=None, filename="data.csv"):
    if directory is None:
        directory = pd.to_datetime("today").strftime("%Y-%m-%d")

    # Buat folder jika belum ada
    os.makedirs(directory, exist_ok=True)

    # Gabungkan directory & filename
    file_path = os.path.join(directory, filename)

    # Simpan DataFrame sebagai CSV
    df.to_csv(file_path, index=False)
    
    print(f"File disimpan di: {file_path}")

# Contoh penggunaan
# save_df_to_csv(df_excel_p, filename="sales_data.csv")
```
